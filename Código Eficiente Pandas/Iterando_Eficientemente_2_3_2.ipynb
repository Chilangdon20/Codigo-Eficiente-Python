# Looping usando la función ``.iterrows()``

## Introducción:

Ahora que sabemos cómo optimizar nuestras elecciones de filas y columnas y sus remplazos de valor,es hora de aprender a iterar através de un DataFrame de oandas de la manera mas eficiente.

El método que presentaremos en esta lección es la función ``.iterrows()``, utilizaremos el conjunto de datos de póker que usamos anteriormente

```
import pandas as pd

data = pd.read_csv("/content/poker_hand.csv")
print(data.head())
```

Recordemos qeu cada fila simboliza la mano de un jugador en un juego de póker que consta de cinco cartas.En cada columna , tenemos el simbolo yel rango para cada uno de las cinco cartas simbolizadas por S & R de la tarjeta respectiva.

Por ejemplo en primer lugar tenemos un 10 de diamantes un Jack , un rey de palos , un 4 de espadas y un as de corazones.


Antes de hablar de como usar la función ``iterroes``vamos a actualizar la noción de función generadora.Los generadores son una herramienta simple para crear iteradores.


Dentro del cuerpo de un generador en lugar de declaraciones de retorno, econtraremos solo declaraciones de rendimiento,puede haber solo una o varias declaraciones de rendimiento.

Aqui podemos ver un generador ``ciudad_nombre_generador()``que produce cuatro nombres de ciudades.Asignamos el generador a la variable ``nombres_ciudades``por simplicidad.

Para acceder a los elemntos que produce un generador , utilizamos la funcion ``next`` de Python.Cada vez que usamos el comando ``next``, el generador producirá el siguiente valor a ceder hasta que nohaya mas valores para ceder.



```
def ciudad_nombre_generador():
  yield('CDMX')
  yield('Playa del Carmen')
  yield('London')
  yield('Tokio')
  yield('Chicago')

nombres_ciudades = ciudad_nombre_generador()
```

En nuestro ejemplo, podemos llamar  a ``next``una vez para que podamos imprimir CDMX en nuestra pantalla, luego la segunda vez imprimir londres y asi sucesivamente.

La función `ìterrows`` es una propiedad de todos los Pnadas DataFrame,cuando se llama produce una lista con 2 elementos.Usaremos este generador para recorrer cada línea de nuestro DataFrame de Poker.


```
gen = data.iterrows()
primer_elemento = next(gen)
```

El primer elemento es el indice de la fila, mientras que el segundo elemento contiene una Serie de pandas de cada caracteristica de la fila, el simbolo y el rango para cada una de las cinco cartas.

Es muy similar a la noción de la función ``enumerate()``, que cuando se aplica  auna lista, devuelve cada elemento junto con su indice.


```
primer_elemento[1]
```

La forma mas intuitiva de iterar sobre un Pandas DataFrame es usar la funcion range que amenudo se llama crud loopiing o looping crudo.

```
import time
start_time = time.time()
for index,values in range(data.shape[0]):
  next
print("Tiempo usando range() : {} segundos ".format(time.time()-start_time))
```

Una forma más inteligente de iterar atraves de un DataFrame es usar la función `ìterrows``que esta optimizada para esta tarea.

```
generador_data = data.iterrows()
start_time = time.time()
for index,values in generador_data:
  next
print("Tiempo usado .iterrows() es: {} segundos".format(time.time()-start_time))
```

Simplemente definimos el ciclo for con dos iteradores, uno para cada fila y otro para todos los valores.Dentro del bucle el comando next indica que el bucle se mueve al siguiente valor del iterador , sin realmente hacer algo.

Tambien podemor notar que el uso de ``iterrows()`` no mejora la velocidad de iteración a través de un DataFrame de pandas.Sin embargo es muy util cuando necesitams una forma mas limpia de usar los valores de cada fila mientras iteramos por el conjunto de datos.


## Looping usando la funcion apply()

Ahora usaremos la funcion ``apply``para realizar una tarea especifica, mientras itera atraves de un DataFrame de Pandas.Seguiremos usando el conjunto de datos de Póker.

La función ``apply`` hace exactamente lo que dice, aplica otra función a todo el DataFrame.La sintaxis de la funcion ``apply`` es simple , mapeamos , utilizando una función lambda en este caso y luego declaramosla funcion que queremos aplicar a cada celda.

A continuacion declararemos la funcion de raiz cuadrada a cada celda del DataFrame.


```
import numpy as np
data_sqrt = data.apply(lambda x: np.sqrt)
print(data_sqrt)
```

En terminos de velocidad coincide con la velocida de usar solo la funcion numpy sqrt en todo el DataFrame.

```
data_sqrt_2 = np.sqrt(data)
data_sqrt_2
```

Pero ¿que sucede cuyando la función de interes toma más de una celda como entrada?

Por ejemplo ¿que pasa si queremos calcular la suma del rango de todas las cartas en cada mano?

En este caso utilizaremos la función ``apply`` que funciona de la misma forma que antes , pero necesitamos agrergar , ``axis =1`` para especificar que lo aplicamos a todas las filas.




```
import time
apply_start_time = time.time()
data[['R1','R2','R3','R4','R5']].apply(lambda x:sum(x),axis = 1)
print("El tiempo usando .apply() es: {}".format(time.time()-apply_start_time))
```

Ahora usaremos la función iterrows y compararemos su tiempo.

```
start_time = time.time()
for ind,value in data.iterrows():
  sum([value[1],value[3],value[5],value[7],value[9]])
print("Tiempo usando .iterrows(): {} ".format(time.time()-start_time))
```

Podemos ver que la función ``iterrows``es mucho mas tardada que ``apply()``

## Vectorizando Pandas series.

En esta seccion aprenderemos otro metodo para iterar sobre un DataFrame y aplicar funciónes especificas,recordemos que una Serie de Pandas  es una matriz etiquetada unidimensional capaz de contener datos de cualquier tipo (entero, cadena, flotante, objetos de python, etc.). Las etiquetas del eje se denominan colectivamente índice. Pandas Series no es más que una columna en una hoja de Excel.

Para comprender cómo podemos reducirla cantidad de iteraciones realizadas por la función recordemos que las unidades fundamentales de Pandas,DataFrames y Series se basan en matrices.

Pandas se desempeñan mas eficientemente cuando una operación es realizada a una matriz completa que a cada valor por separado o secuencialmente.La vectorizació, es el proceso de ejecución de operaciones en matrices enteras.


Mostraremos cómo realizar la vectorización de pandas utilizando el conjunto de datos de póker.Nuevamente queremos calcular la suma de los rangos de todas las cartas en cada mano.

Para hacer esto dividimos el conjunto de datos de póker manteniendo solo las columnas que contienen los rangos de cada carta.

Luego llamamos a la propiedad ``sum()`` del DataFrame y especificamos que queremos la suma para cada fila usando ``axis=1``.

```
start_time = time.time()
data[['R1','R2','R3','R4','R5']].sum(axis=1)
print("El tiempo usando la vectorización de pandas es : {} segundos".format(time.time()-start_time))
```

Al final , imprimimos la suma de las primeras cinco filas de los datos,nuestro objetivo es encontrar el método más eficiente para realizar esta tarea de la manera mas efciciente posible.




```
data[['R1','R2','R3','R4','R5']].sum(axis=1).head()
```

Para cerrar esta sección , discutirtemos otro método de vectorización para iterar eficientemente sobre un DataFrame de pandas y aplicando una función especifica usando NumPy.


La libreria NumPy se define como un paquete fundamental para cientificos de informatica en Python , ya que realiza operaciones ocultas en código C optimizado y precompilado.

Similar a Pandas que trabaja con matrices, NumPy opera en matrices llamadas *ndarays*.

```
import numpy as np
df = pd.DataFrame(
    {'Col1':[0,1,2,3,4,5,6]},dtype=np.int8)
print(df)
```

Una diferencia importante entre Series y ndarrays es que se omiten muchas operaciones como la indexación, la verificación del tipo de datos, etc.

Como resultado , las operaciones en matrices NumPy pueden ser significativamente más rápidas que las operaciones en la serie de Pandas.


## ¿Cuando usar matrices NumPy?

Se pueden usar en lugar de la serie de Pandas cuando,la funcionalidad que ofrece la serie pandas no es crítica.

Para los problemas que exploramos en esta seccion, podríamos usar NumPy ndarrays en lugar de series de Pandas.La pregunta seria... ¿ Esto seria mas eficiente?
A continuación veremos cómo realizar la vectorización de NumPy utilizando el mismo conjunto de datos póker que en las lecciones anteriores.



```
np = np.array(range(7))
print(np)
```

Nuevamente queremos calcular la suma de los rangos de todas las cartas en cada mano.Convertimos nuestras matrices de rango de series pandas a matrices NumPy simplemente usando el metodo ``values``de la seria de pandas que devuelve una serie de pandas como una mátriz numérica NumPy.

```
s_t = time.time()
data[['R1','R2','R3','R4','R5']].values.sum(axis =1)
print("El tiempo usando la vecotrización Numpy es : {}".format(time.time()-s_t))
```
