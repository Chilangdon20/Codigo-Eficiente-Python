Cuando creamos una clase, podemos elegir que tenga variables estáticas, esto significa que no cambian,no importa lo que hagamos a los miembros de la clase.

En la siguiente clase, tenemos una clase llamada `Tiburon` nuetra variable estática son los ojos, no importa qué tipo de tiburon sea , esa figura siempre tendra dos ojos y esa variable continuará através de todas las iteraciones de la clase.



```
# Clase Tiburon.
class Tiburon():

  ojos = 2
  def __init__(self,dientes):
    self.dientes = dientes
```

Los dientes , podemos establecerlo y podemos establecerlo cuando contruimos la clase, esto es lo que se conoce como una variable de instancia.

En este ejemplo , creamos una instancia de tiburon llamada Megalodon, que tiene 90 dientes.
Podemos ver que si imprimimos dientes, nuetra variable de instancia es 90.

```
# Construyendo un tiburon
megalodon = Tiburon(90)
megalodon.Tiburon = 90
megalodon.dientes
```

S imimprimimos ojos vemos que es 2.

```
megalodon.ojos
```

Ahora, podemos establecer dientes de manera diferente para diferentes tiburones, creemos por ejemplo un toburon bebe , que solo tiene 15 dientes.

Cuando llamamos dientes, vemos que la respuesta es 15.

```
mako = Tiburon(15)
mako.Tiburon = 15
mako.dientes 
```

Si llamamos ojos, podemos ver que la respuesta sigue siendo dos, apesar de que no la llamamos para instanciar esa clase.

```
mako.ojos
```
