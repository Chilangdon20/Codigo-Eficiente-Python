# Métodos en Clases.

Introducción:

Una función es un fragmento de código que realiza una acción en otro fragmento de código,esta sección cubriremos como funciónan los métodos o funciones dentro de las clases.
Veamos nuestra clase DataS:

```
import numpy as np
```

```
class DataS:

  # método init
  def __init__(self,filename):
    self.filename = filename

  # crear: data_shell
  def create_datashell(self):
    self.array = np.genfromtxt(self.filename, delimiter=',',dtype=None)
    return self.array

  # crear: rename_column
  def rename_column(self,old_colname,new_colname):
    for index,value in enumerate(self.array[0]):
      if value == old_colname.encode('UTF-8'):
        self.array[0][index] = new_colname
    return self.array
```

Podemos ver que una clase puede tener muchos metodos diferentes, este tiene dos especificamente:

* `create_datashell` - Inicializa un DataFrame apartir de su nombre de archivo.


* `rename_column`- Cambia el nombre de una columna en función de un nombre de columna antiguo y un nuevo nombre de columna.


Cada uno de estos métodos deviuelve variables, que se devuelven a la clase como un atributo de clase.

¿Cómo funciona todo esto?

También devolvemos algo, una matriz NumPy, por lo general , cuando trabajamos con una función, queremos imprimir los resultados o canalizrlos a otra función.

Dentro de una clase, generalmente devolvemos un elemento , por que queremos trabajar con ese elemento en otras partes de la clase, en lugar de imprimirlo.
